```python
import pandas as pd

# Memuat file CSV
file_path = 'UGUISU.csv'
df = pd.read_csv(file_path)

# Menampilkan nama kolom untuk memastikan kolom yang digunakan benar
print(df.columns)

# Menghilangkan spasi pada nama kolom jika ada
df.columns = df.columns.str.strip()

# Regex pattern untuk mengekstrak angka menit (satu atau lebih digit)
regex_pattern = r'(\d+)'

# Menggunakan regex untuk mengekstrak angka menit dari kolom "Time Stamp"
df['Minutes_Extracted'] = df['Time Stamp'].astype(str).str.extract(regex_pattern)

# Mengubah hasil ekstraksi ke tipe numerik
df['Minutes_Extracted'] = pd.to_numeric(df['Minutes_Extracted'], errors='coerce')

# Menampilkan beberapa baris pertama hasil ekstraksi
print(df[['Time Stamp', 'Minutes_Extracted']].head())
```

##### <div style="text-align: justify;">Kode ini digunakan untuk membersihkan dan mengekstrak data dari kolom Time Stamp dalam dataset. Langkah pertama adalah memastikan bahwa format kolom Time Stamp konsisten dan dapat diproses dengan benar. Dengan menghapus spasi yang tidak perlu dan mengekstrak angka menit menggunakan regular expression (regex), kode ini mengurangi masalah yang bisa muncul jika data waktu tidak terstruktur dengan baik. Hal ini penting untuk analisis lebih lanjut, seperti pengelompokan atau perhitungan berbasis waktu.</div>
##### <div style="text-align: justify;">Proses ekstraksi memastikan hanya angka yang diambil dari kolom Time Stamp, yang membantu menghindari data yang tidak sesuai, seperti teks atau simbol yang tercampur. Setelah itu, data yang diekstrak diubah menjadi tipe numerik agar bisa digunakan dalam perhitungan lebih lanjut. Jika ada nilai yang tidak bisa dikonversi, nilai tersebut diubah menjadi NaN, yang memudahkan untuk mendeteksi dan memperbaiki data yang bermasalah. Hasilnya, kolom Minutes_Extracted yang berisi angka yang terstruktur siap digunakan untuk analisis berbasis waktu atau pengolahan data lebih lanjut.</div>

#### 5. Membersihkan Kolom dengan Kombinasi metode str dengan NumPy

#####  <div style="text-align: justify;">Fungsi membersihkan kolom dengan menggabungkan metode str dan NumPy dalam proses pembersihan data bertujuan untuk memastikan data dalam kolom yang berisi teks atau string menjadi lebih terstruktur, konsisten, dan mudah diproses. Dengan menggunakan metode str, kita dapat memperbaiki format teks, seperti menghapus spasi berlebih, mengganti nilai yang tidak konsisten, atau mengekstrak bagian tertentu dari teks. Sementara itu, NumPy berfungsi untuk menangani nilai yang hilang, mengubah tipe data yang tidak sesuai, atau mengganti data yang rusak dengan nilai yang lebih tepat. Kombinasi kedua metode ini memungkinkan kita membersihkan data secara efisien dan cepat, sehingga data siap untuk analisis lebih lanjut tanpa masalah terkait format atau kualitas.</div>

```python
file_path = 'NEPALISAT.csv'
df = pd.read_csv(file_path)
pub = df['Tbatt raw)']
AF = pub.str.contains('AF')
AF[:5]
yoi = pub.str.contains('AF')
df['Tbatt raw)'] = np.where(AF, 'B0',
                                      np.where(yoi, 'yoi',
                                               pub.str.replace('-', ' ')))
df['Tbatt raw)'].head()
```

##### <div style="text-align: justify;">Fungsi kode ini bertujuan untuk membersihkan dan memperbaiki nilai dalam kolom 'Tbatt raw)' pada dataset dengan mengganti beberapa nilai yang tidak konsisten atau salah format. Salah satu langkah yang dilakukan adalah mengganti nilai yang mengandung substring 'AF' dengan nilai yang lebih jelas dan konsisten, seperti 'B0' atau 'yoi'. Hal ini berguna untuk memperbaiki kesalahan penulisan atau untuk menstandarkan data, sehingga data yang ada menjadi lebih mudah dipahami dan digunakan. Selain itu, tanda minus ('-') yang mungkin tidak sesuai atau tidak relevan dalam kolom tersebut digantikan dengan spasi (' '), yang membantu menjaga konsistensi format data.</div>
##### <div style="text-align: justify;">Kode ini juga menggunakan metode str.contains() untuk mendeteksi pola teks tertentu dalam kolom dan mengganti nilai berdasarkan pola yang ditemukan. Ini memungkinkan untuk membersihkan data dengan menyesuaikan nilai yang tidak sesuai atau tidak relevan. Dengan melakukan penggantian nilai secara sistematis dan berdasarkan kondisi yang jelas, data menjadi lebih konsisten, lebih terstruktur, dan lebih siap untuk analisis lebih lanjut. #### 6. Membersihkan Seluruh Dataset dengan Fungsi applymap
---

##### <div style="text-align: justify;">Fungsi applymap() dalam pandas digunakan untuk menerapkan sebuah fungsi atau operasi pada setiap elemen dalam seluruh DataFrame secara bersamaan. Hal ini sangat berguna dalam proses pembersihan data karena kita dapat membersihkan atau mentransformasi data di seluruh dataset dengan cepat dan efisien, tanpa perlu memproses kolom atau baris satu per satu. Misalnya, kita bisa melakukan perubahan seperti menghapus spasi ekstra, mengganti nilai tertentu, atau mengubah format data di setiap sel, semuanya dalam satu langkah. Dengan applymap(), kita bisa memastikan bahwa perubahan yang diterapkan konsisten di seluruh data, memudahkan pembersihan, dan menghemat waktu dalam proses pembersihan data.</div>

```python
# Initialize an empty list to store the processed data
processed_data = []
file_path = 'NEPALISAT.csv'
# Open the file and read line by line
with open(file_path, 'r') as file:
    for line in file:
        line = line.strip()  # Remove extra whitespace and newline characters
        if '[edit]' in line:  # Check if the line indicates a category
            category = line  # Update the category
        else:
            # Append the category and current line as a tuple
            processed_data.append((category, line))

# Print the first 5 processed items
processed_data[:5]
```

```python
import pandas as pd
nepalisat_df = pd.DataFrame(processed_data,
                        columns=['category', 'line'])

nepalisat_df.head()
```

```python
def get_citystate(item):
    if pd.isna(item):  # Periksa jika item adalah NaN atau None
        return item  # Kembalikan item tanpa perubahan
    item = str(item)  # Pastikan item berupa string
    if ' (' in item:
        return item[:item.find(' (')].strip()
    elif '[' in item:
        return item[:item.find('[')].strip()
    else:
        return item.strip()

# Terapkan fungsi ke seluruh DataFrame
nepalisat_df_cleaned = nepalisat_df.applymap(get_citystate)

# Menampilkan 5 baris pertama dari DataFrame yang telah dibersihkan
nepalisat_df_cleaned.head()
```
