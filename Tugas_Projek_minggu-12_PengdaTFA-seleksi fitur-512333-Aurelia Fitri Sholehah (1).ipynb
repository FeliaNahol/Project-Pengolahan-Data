### 1. Seleksi Unvariate

##### Seleksi Univariate adalah metode dalam analisis data yang digunakan untuk memilih fitur atau variabel tunggal (univariate) yang paling relevan atau penting dari sekumpulan data untuk tujuan tertentu, seperti analisis atau model prediktif. Dalam seleksi univariate, setiap fitur dianalisis secara terpisah untuk melihat seberapa besar kontribusinya terhadap target atau variabel yang ingin diprediksi, menggunakan teknik statistik seperti uji t, chi-square, atau korelasi Pearson.
##### Dalam konteks artikel tentang dataset on-orbit dari sistem kelistrikan (EPS) pada CubeSats BIRDS, seleksi univariate bisa digunakan untuk menganalisis variabel seperti tegangan (mV), arus (mA), dan suhu (Celsius) untuk menentukan fitur mana yang paling berpengaruh dalam memprediksi keadaan normal atau anomali sistem kelistrikan satelit. Misalnya, jika kita ingin memprediksi kegagalan panel surya atau masalah pada baterai, kita bisa menggunakan seleksi univariate untuk memilih apakah tegangan atau arus yang dihasilkan oleh panel surya lebih relevan dalam mengidentifikasi kegagalan tersebut. Dengan memilih variabel yang paling signifikan, kita bisa memperbaiki model prediktif untuk deteksi dini masalah pada satelit.
##### Selanjutnya, dalam penelitian ini, yang mencakup data dari beberapa CubeSat yang beroperasi di orbit, seleksi univariate membantu untuk fokus pada fitur yang paling berpengaruh dalam klasifikasi anomali. Sebagai contoh, jika data menunjukkan korelasi yang tinggi antara suhu dan arus pada satelit dengan kegagalan panel surya, maka suhu mungkin menjadi fitur penting untuk dipertimbangkan dalam model diagnosa kesalahan. Dengan menggunakan seleksi univariate, peneliti dapat meminimalkan kompleksitas model dan meningkatkan efisiensi dalam mendeteksi potensi kegagalan sistem kelistrikan berdasarkan data yang tersedia.

##### - Satelite UGUISU

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder

# Load your dataset
data = pd.read_csv('UGUISU.csv')

# Define independent and target variables
X = data.iloc[:, 0:20]  # independent columns
y = data.iloc[:, -1]     # target column i.e. price range

# Convert target variable to categorical if it is not already
if y.dtype != 'object':
    # You can bin continuous values into categories if needed
    y = pd.qcut(y, q=5, labels=False)  # Example: Convert to 5 categories

# If y is still not categorical, use LabelEncoder
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Convert categorical features in X to numeric using one-hot encoding
X = pd.get_dummies(X, drop_first=True)

# Check for negative values in X
print("Checking for negative values in X:")
print(X[X < 0].count())  # Count of negative values in each column

# If there are negative values, you can handle them as follows:
# Option 1: Replace negative values with 0
X[X < 0] = 0

# Option 2: Alternatively, take the absolute value (if it makes sense for your data)
# X = X.abs()

# Now try fitting the SelectKBest again
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X, y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

# Concatenate two dataframes for better visualization
featureScores = pd.concat([dfcolumns, dfscores], axis=1)
featureScores.columns = ['Specs', 'Score']  # naming the dataframe columns
print(featureScores.nlargest(10, 'Score'))  # print 10 best features
```

##### Hasil seleksi univariate pada dataset yang diberikan menunjukkan bahwa proses seleksi fitur telah dilakukan untuk memilih variabel-variabel yang paling relevan dalam memprediksi variabel target atau dalam analisis klasifikasi. Dalam seleksi univariate, setiap fitur dievaluasi secara individual untuk menentukan seberapa besar kontribusinya terhadap prediksi target, yang dalam hal ini bisa merujuk pada status sistem kelistrikan satelit, seperti kondisi normal atau adanya anomali. Fitur dengan skor yang lebih tinggi dianggap memiliki pengaruh yang lebih besar terhadap hasil prediksi, sementara fitur dengan skor yang lebih rendah dianggap kurang signifikan.
##### Sebagai contoh, fitur Vmx (mV) dengan skor 765918.242372 berada di urutan teratas, yang menunjukkan bahwa variabel ini memiliki relevansi yang tinggi dalam membedakan kelas atau status target, seperti kondisi normal atau anomali pada sistem kelistrikan satelit. Skor tinggi pada fitur ini mengindikasikan bahwa perubahan pada nilai Vmx memiliki pengaruh besar dalam analisis, menjadikannya fitur yang sangat penting untuk dipertimbangkan dalam model prediksi. Di sisi lain, fitur seperti Tmx (°C) yang memiliki skor 887.411219 menunjukkan bahwa variabel ini kurang relevan dalam membedakan atau memprediksi target, sehingga mungkin tidak perlu menjadi fokus utama dalam analisis atau model.
##### Secara keseluruhan, seleksi univariate berfungsi untuk mengidentifikasi fitur-fitur yang paling berpengaruh atau relevan dalam analisis atau klasifikasi data. Dengan menggunakan skor dari seleksi univariate, kita dapat lebih efisien dalam memilih fitur yang akan digunakan untuk model prediksi. Fitur dengan skor lebih tinggi, seperti Vmx (mV) dan Vpx (mV), cenderung lebih informatif dan akan lebih sering digunakan dalam model analisis, karena mereka menunjukkan hubungan yang kuat dengan variabel target. ##### Selanjutnya, seleksi fitur penting dapat membantu para pengembang satelit dan peneliti untuk membandingkan performa sistem kelistrikan di berbagai skenario operasional dan menyesuaikan misi mereka dengan lebih efisien, sesuai dengan daya yang tersedia di orbit.
```python
import pandas as pd
import numpy as np

data = pd.read_csv("UGUISU.csv")
X = data.iloc[:, 0:20]  # independent columns
y = data.iloc[:, -1]     # target column i.e. price range

# Convert target variable to categorical if it is not already
if y.dtype != 'object':
    # You can bin continuous values into categories if needed
    y = pd.qcut(y, q=5, labels=False)  # Example: Convert to 5 categories

# If y is still not categorical, use LabelEncoder
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Convert categorical features in X to numeric using one-hot encoding
X = pd.get_dummies(X, drop_first=True)

# Check for negative values in X
print("Checking for negative values in X:")
print(X[X < 0].count())  # Count of negative values in each column

# If there are negative values, you can handle them as follows:
# Option 1: Replace negative values with 0
X[X < 0] = 0

from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,y)

print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers

#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

Selain itu, nilai untuk variabel tegangan dan suhu berada dalam rentang yang cukup kecil, yakni antara 0 hingga sekitar 0.04. Rentang yang sempit ini mengindikasikan bahwa pengukuran tegangan dan suhu dilakukan dalam kondisi yang relatif stabil, atau mungkin dalam skala yang kecil, yang menunjukkan pengendalian yang ketat terhadap variabel-variabel tersebut.

### 3. Matriks Korelasi

##### - Satelite UGUISU

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Baca data dari file
data = pd.read_csv("UGUISU.csv")

# Pilih hanya kolom numerik
numeric_data = data.select_dtypes(include=[np.number])

# Hitung matriks korelasi
corrmat = numeric_data.corr()

# Urutkan fitur berdasarkan korelasi
top_corr_features = corrmat.index

# Plot heatmap
plt.figure(figsize=(20, 20))
g = sns.heatmap(numeric_data[top_corr_features].corr(), annot=True, cmap="RdYlGn")

plt.show()
```
