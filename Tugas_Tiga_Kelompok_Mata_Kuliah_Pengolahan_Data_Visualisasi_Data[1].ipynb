## 1. Deskripsi Dataset 
<div style="text-align: justify;">Kumpulan data tersebut  berisi informasi tentang efisiensi energi bangunan di Kota New York, berdasarkan Undang-Undang Lokal 33 (LL33) kota New York pada tahun 2021. Undang-undang tersebut mengharuskan pemilik bangunan besar untuk mengukur dan mengungkapkan konsumsi energi dan airs etiap tahun. Data ini dikumpulkan oleh Departemen Keuangan Kota New York. Data terdiri atas beberapa karakteristik setiap bangunan yaitu :</div>

- 10 Digit BBL (Borough, Block, Lot): Pengidentifikasi unik untuk setiap properti di NYC
- Nomor Jalan dan Nama Jalan: Alamat bangunan
- DOF Gross Square Footage: Luas lantai kotor bangunan sebagaimana dicatat oleh Departemen Keuangan
- Nilai Efisiensi Energi: Nilai huruf dari A-D yang ditetapkan berdasarkan skor efisiensi energi bangunan

## 3. Visualisasi Data

##### - Bar Chart

<div style="text-align: justify;">Bar chart adalah jenis grafik yang menggunakan batang atau bilah untuk mewakili data. Setiap batang menunjukkan nilai atau frekuensi dari suatu kategori atau variabel. Biasanya, panjang atau tinggi batang menunjukkan nilai numerik dari data, sehingga semakin panjang batangnya, semakin besar nilai yang diwakilinya

```python
#Memanggil pustaka pandas, matplotlib, dan seaborn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Membaca dataset dalam format csv dan diubah ke DataFrame
df = pd.read_csv('energy_disclosure_2021_rows.csv')

#Melakukan ekstrasi pada kolom Energy Star 1-100 Score dan Street Name
energy_score = df['Energy_Star_1-100_Score'].head(10).tolist()
efficiency_grade = df['Street_Name'].head(10).tolist()

# Mengurutkan DataFrame berdasarkan nilai pada kolom Energy Star 1-100 Score
df_sorted = df.sort_values(by='Energy_Star_1-100_Score').head(10)

#Menampilkan bar chart
plt.bar(range(len(df_sorted)), df_sorted['Energy_Star_1-100_Score'], tick_label=df_sorted['Street_Name'].tolist())
plt.xticks(rotation=90)
plt.ylabel('Energy Star Score')
plt.title('Street Name')
plt.show()
```

<div style="text-align: justify;">Berdasarkan grafik batang yang ditampilkan, terlihat perbedaan nilai Energy Star Score dari berbagai jalan atau area di suatu wilayah. Nilai ini mencerminkan seberapa efisien suatu bangunan atau area dalam penggunaan energi. Dari grafik, terdapat 10 lokasi yang dibandingkan. Area dengan nilai terendah adalah North End Avenue dan South William Street, keduanya memiliki skor sekitar 2-3. Sedangkan, Maiden Lane memiliki skor tertinggi, yaitu sekitar 17,5. Hal ini menunjukkan bahwa Maiden Lane memiliki efisiensi energi yang lebih baik dibandingkan dengan lokasi lainnya.
<br>
<div style="text-align: justify;">Secara umum, tren menunjukkan peningkatan nilai Energy Star Score dari kiri ke kanan, mengindikasikan bahwa ada variasi signifikan dalam efisiensi energi di antara lokasi yang diukur. Area dengan nilai skor di bawah 10, seperti Broad Street, Battery Place, dan Beaver Street, menunjukkan kinerja efisiensi energi yang relatif rendah. Sementara itu, area seperti Wall Street, Rector Street, dan Maiden Lane menunjukkan performa yang lebih baik, dengan nilai di atas 15. Hal ini dapat menunjukkan perbedaan dalam manajemen energi atau infrastruktur bangunan di tiap lokasi yang memengaruhi skor efisiensi energi.

##### - Histogram

<div style="text-align: justify;">Histogram adalah sebuah grafik yang digunakan untuk menampilkan distribusi frekuensi dari suatu data. Histogram terdiri dari serangkaian batang (bars) yang masing-masing mewakili kelompok data atau kelas interval dengan tinggi batang menunjukkan frekuensi atau jumlah data yang terdapat pada kelas interval tersebut. Dengan histogram, dapat dengan mudah dilihat bagaimana suatu data terdistribusi dan mengidentifikasi pola seperti apakah data cenderung berkumpul di satu titik atau tersebar merata.

```python
#Memanggil pustaka pandas dan matplotlib
import pandas as pd
import matplotlib.pyplot as plt
# Membaca dataset dalam format csv dan diubah ke DataFrame khusunya untuk kolom Energy Star 1-100 Score
df = pd.read_csv("energy_disclosure_2021_rows.csv", usecols=['Energy_Star_1-100_Score'])
#Menampilkan histogram
plt.hist(df)
plt.title("Histogram")
plt.show() 
```

<div style="text-align: justify;">Histogram yang ditampilkan menggambarkan distribusi frekuensi suatu data numerik dalam rentang 0 hingga 100. Dari histogram ini, terlihat bahwa data memiliki sebaran yang tidak merata dan terdapat kecenderungan tertentu dalam distribusinya. Pada sumbu horizontal (x-axis), terdapat interval kelas yang berkisar dari 0 hingga 100, dan sumbu vertikal (y-axis) menunjukkan jumlah atau frekuensi data pada tiap interval tersebut.

<div style="text-align: justify;">Data yang ditampilkan dalam histogram ini menunjukkan bahwa banyak bangunan memiliki skor Energy Star yang tinggi, yaitu mendekati 80 hingga 100. Ini mengindikasikan bahwa kebanyakan bangunan dalam dataset memiliki performa efisiensi energi yang baik. Sementara itu, bangunan dengan skor rendah (di bawah 20) relatif lebih sedikit, menunjukkan adanya potensi untuk perbaikan efisiensi energi pada sebagian kecil bangunan. 

<div style="text-align: justify;">Distribusi data ini left-skewed atau memiliki kecenderungan distribusi condong ke kiri. Artinya, lebih banyak bangunan yang memiliki skor tinggi dibandingkan yang memiliki skor rendah.

##### - Scatter Plot

<div style="text-align: justify;">Scatter plot atau diagram pencar adalah jenis grafik yang digunakan untuk menampilkan hubungan atau keterkaitan antara dua variabel kuantitatif. Pada scatter plot, setiap titik pada grafik mewakili satu pasangan data dari dua variabel yang diukur.

```python
#Memanggil pustaka pandas dan seaborn
import seaborn as sns
import pandas as pd
# Membaca dataset dalam format csv dan diubah ke DataFrame
df = pd.read_csv("energy_disclosure_2021_rows.csv")
#Menampilkan Scatter Plot
sns.scatterplot(x=df['Street_Number'], y=df['Energy_Star_1-100_Score'])
```

<div style="text-align: justify;">Berdasarkan scatter plot yang ditampilkan, kita dapat melihat hubungan antara Street Number (nomor jalan) dan Energy Star 1-100 Score (skor efisiensi energi). Pada sumbu X terdapat nomor jalan, sementara sumbu Y menunjukkan skor Energy Star dari 1 hingga 100. Secara visual, titik-titik data tersebar tanpa menunjukkan pola yang jelas atau korelasi linear yang kuat. Hal ini mengindikasikan bahwa nomor jalan tidak memiliki hubungan langsung yang signifikan dengan skor efisiensi energi suatu bangunan. Terlihat bahwa skor efisiensi bervariasi di sepanjang berbagai nomor jalan, dari angka rendah hingga tinggi, tanpa adanya kecenderungan kenaikan atau penurunan yang konsisten.

<div style="text-align: justify;">Sebagian besar titik data tampak terkumpul di rentang skor efisiensi antara 0 hingga 100, dengan kepadatan yang lebih tinggi di sekitar nilai 80-100. Ini menunjukkan bahwa beberapa bangunan memiliki efisiensi energi yang cukup tinggi terlepas dari nomor jalannya. Namun, masih terdapat sejumlah bangunan dengan skor rendah atau sedang yang tersebar pada berbagai nomor jalan. Secara keseluruhan, scatter plot ini menunjukkan bahwa faktor lokasi yang diwakili oleh nomor jalan mungkin bukan merupakan indikator yang signifikan untuk menentukan skor Energy Star suatu bangunan, sehingga perlu dianalisis lebih lanjut faktor-faktor lain yang lebih relevan.

##### - Line Plot

<div style="text-align: justify;">Line plot atau diagram garis adalah jenis grafik yang menampilkan titik data di sepanjang garis bilangan, dengan nilai-nilai yang digambarkan sebagai titik atau tanda. Titik-titik ini kemudian dihubungkan dengan garis untuk menunjukkan tren atau perubahan data seiring waktu atau dalam kaitannya dengan variabel lainnya. Diagram garis sering digunakan untuk menggambarkan bagaimana sesuatu berubah atau berkembang, sehingga cocok untuk melacak data yang bersifat kontinu.

```python
#Memanggil pustaka pandas dan matplotlib, d
import matplotlib.pyplot as plt
import pandas as pd
# Membaca dataset dalam format csv dan diubah ke DataFrame
df = pd.read_csv("energy_disclosure_2021_rows.csv", usecols=['Energy_Efficiency_Grade', 'Energy_Star_1-100_Score'])

# Memasukkan data frame Energy Effiiciency Grade ke variabel x dan data frame Energy Star 1-100 ke variabel y
x=df['Energy_Efficiency_Grade']
y=df['Energy_Star_1-100_Score']

#Melakukan plotting data
plt.plot(x, y)
plt.title("Plotting Dataset")
plt.ylabel('Energy Star 1-100 Score')
plt.xlabel('Ennergy Efficiency Grade')
plt.show()
```

<div style="text-align: justify;">Grafik ini menunjukkan hubungan antara Energy Efficiency Grade (A, B, C, dan D) dengan Energy Star 1-100 Score. Tampak bahwa distribusi skor berada pada kisaran yang luas untuk setiap kategori efisiensi energi, meskipun terdapat kecenderungan umum. Untuk grade A, skor cenderung lebih tinggi, dengan banyak titik data berada di kisaran 80 hingga 100. Hal ini menunjukkan bahwa bangunan atau entitas dengan grade A umumnya memiliki skor Energy Star yang lebih tinggi, yang mengindikasikan efisiensi energi yang lebih baik.

<div style="text-align: justify;">Sebaliknya, untuk grade yang lebih rendah seperti D, skor cenderung menurun, dengan lebih banyak data yang tersebar di kisaran skor lebih rendah (20-60). Pola ini menunjukkan bahwa semakin rendah grade efisiensi energi, skor Energy Star juga cenderung menurun, mengindikasikan performa energi yang kurang efisien. Namun, terdapat tumpang tindih di beberapa kategori, terutama antara grade B dan C, yang mungkin menunjukkan bahwa tidak semua bangunan dengan grade yang sama memiliki performa yang seragam dalam hal efisiensi energi.

##### - Regresi

<div style="text-align: justify;">Regresi plot atau plot regresi adalah grafik yang digunakan untuk menunjukkan hubungan antara dua variabel, di mana salah satu variabel dianggap sebagai variabel independen (X) dan variabel lainnya sebagai variabel dependen (Y). Grafik ini digunakan untuk menganalisis pola atau tren dalam data dan untuk membuat prediksi berdasarkan hubungan antara kedua variabel tersebut.

```python
#Memanggil pustaka seaborn
import seaborn  as sns
# Membaca dataset dalam format csv dan diubah ke DataFrame
df = pd.read_csv('energy_disclosure_2021_rows.csv')
#Menampilkan grafik regresi linear
sns.regplot(x="DOF_Gross_Square_Footage", y="Energy_Star_1-100_Score", data=df) 
plt.ylim(0,)
```

<div style="text-align: justify;">Grafik ini adalah scatter plot dengan garis regresi yang menunjukkan hubungan antara DOF Gross Square Footage (luas bangunan dalam kaki persegi) dan Energy Star 1-100 Score pada suatu dataset bangunan. Titik-titik pada grafik mewakili bangunan individu, dengan sumbu x menunjukkan luas bangunan dan sumbu y menunjukkan skor Energy Star, yang mengukur efisiensi energi. Dari visualisasi ini, terdapat kecenderungan bahwa bangunan dengan luas yang lebih besar cenderung memiliki skor Energy Star yang sedikit lebih tinggi, meskipun korelasinya terlihat lemah. Hal ini ditunjukkan oleh garis regresi yang memiliki kemiringan positif tetapi dengan banyak titik yang tersebar secara acak di sekitar garis.

<div style="text-align: justify;">Meskipun garis regresi menunjukkan tren positif, rentang data yang luas dan penyebaran titik-titik menunjukkan variabilitas yang tinggi dalam skor efisiensi energi pada semua ukuran bangunan. Pada luas yang kecil (di bawah 0,5 juta kaki persegi), terdapat bangunan dengan skor Energy Star yang sangat bervariasi, dari skor rendah hingga tinggi. Di sisi lain, pada bangunan dengan luas yang sangat besar (mendekati 3 juta kaki persegi), skor cenderung berada di kisaran yang lebih tinggi, meskipun masih terdapat outlier. Secara keseluruhan, grafik ini menunjukkan bahwa meskipun luas bangunan mungkin berpengaruh terhadap skor efisiensi energi, ada banyak faktor lain yang memengaruhi skor Energy Star.

##### - Box Plot

<div style="text-align: justify;">Box plot (atau diagram kotak) adalah jenis grafik yang digunakan untuk menggambarkan distribusi data secara visual, menunjukkan rangkuman statistik dari data seperti nilai minimum, kuartil pertama (Q1), median (Q2), kuartil ketiga (Q3), dan nilai maksimum. Box plot sangat berguna untuk menunjukkan sebaran data serta mengidentifikasi outlier (nilai yang jauh berbeda dari data lainnya).

```python
#Memanggil pustaka seaborn
import seaborn  as sns
# Membaca dataset dalam format csv dan diubah ke DataFrame
df = pd.read_csv('energy_disclosure_2021_rows.csv')
#Menampilkan box plot dengan sumbu x yaitu Energy Efficiency Grade dan sumbu y yaitu Energy Star 1-100 Score
ax = sns.boxplot(x="Energy_Efficiency_Grade", y="Energy_Star_1-100_Score", data=df)
```

<div style="text-align: justify;">Boxplot ini menggambarkan distribusi skor Energy Star (skala 1 hingga 100) untuk empat grade efisiensi energi (A, B, C, dan D). Grade A memiliki skor median tertinggi, dengan sebagian besar skor berada di kisaran 80 hingga mendekati 100. Hal ini menunjukkan bahwa bangunan dengan grade A umumnya memiliki efisiensi energi yang tinggi. Grade C berada di bawahnya, dengan median skor sekitar 60, meskipun ada beberapa skor yang jauh lebih rendah sebagai pencilan. Sementara itu, grade B memiliki distribusi yang lebih sempit, dengan sebagian besar skor berkisar di sekitar angka 70-an, yang menunjukkan konsistensi dalam efisiensi energi.

<div style="text-align: justify;">Sebaliknya, Grade D memiliki sebaran skor yang lebih luas, dengan median skor sekitar 50. Hal ini menunjukkan bahwa bangunan dengan grade D cenderung memiliki efisiensi energi yang lebih rendah dan lebih bervariasi. Rentang yang lebar ini mengindikasikan bahwa performa efisiensi energi bangunan dengan grade D tidak konsisten, dengan beberapa bangunan mencapai skor yang cukup tinggi, sementara yang lain sangat rendah. Secara keseluruhan, tren dari boxplot ini menunjukkan bahwa bangunan dengan grade lebih tinggi (A, B) biasanya memiliki skor Energy Star yang lebih baik, sedangkan grade yang lebih rendah (C, D) cenderung memiliki skor yang lebih rendah dan lebih bervariasi.
